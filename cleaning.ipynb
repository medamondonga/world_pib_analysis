```python
import pandas as pd
```

```python
df = pd.read_csv("pib_data.csv")
df.head()
```

## Delete rows with continent value in Pays/Zone column

```python
df=df[~df["Pays/Zone"].isin(["Afrique", 
                           "Europe", 
                           "Asie", 
                           "Amérique du Nord", 
                           "Amérique du Sud", 
                           "Océanie",
                           "Monde",
                           "Union européenne"])]
df = df.drop(columns="Unnamed: 0")
df.info()
```

## Select all value columns, replaces “–” and missing entries with “0”, and returns the cleaned DataFrame.

```python
values = df.columns[1:]
df[values] = df[values].replace("-", "0")
df[values] = df[values].fillna("0")
df.head()
```

## Clean the numeric columns by removing unwanted characters, standardizing commas to dots, converting everything to floats, and then shows the DataFrame’s structure.

```python
df[values] = (
    df[values]
        .replace(r"[^\d,.\-]", "", regex=True)  
        .replace(",", ".", regex=True)          
        .astype(float)
)
df.info()
```

## Standardize the column names by making them lowercase and replacing spaces with underscores, then selects all value columns

```python
df.columns = (df.columns.str.lower().str.replace(" ", "_"))
values = df.columns[1:]
df.head()
```

## Keep only the rows where all numeric values are strictly greater than zero.

```python
df_clean = df[(df[values] > 0).all(axis=1)]
df_clean.head()
```

```python
df_clean.describe() 
```

```python
df_clean.to_csv("pib_data_cleaned.csv", index=False)
```
