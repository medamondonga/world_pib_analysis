```python
import pandas as pd
```

```python
df = pd.read_csv("pib_data.csv")
df.head()
```

```python
#delete rows with continent value in Pays/Zone column
df=df[~df["Pays/Zone"].isin(["Afrique", 
                           "Europe", 
                           "Asie", 
                           "Amérique du Nord", 
                           "Amérique du Sud", 
                           "Océanie"])]
df.info()
```

```python
values = df.columns[2:]
df[values] = df[values].replace("-", "0")
df[values] = df[values].fillna("0")
df.head()
```

```python
df[values] = (
    df[values]
        .replace(r"[^\d,.\-]", "", regex=True)   # garder seulement 0-9 , . -
        .replace(",", ".", regex=True)           # virgule -> point
        .astype(float)
)
df.info()
```

```python
df.columns = (df.columns.str.lower().str.replace(" ", "_"))
values = df.columns[2:]
df.head()
```

```python
df_clean = df[(df[values] > 0).all(axis=1)]
df_clean.head()
```

```python
df_clean.to_csv("pib_data_cleaned.csv", index=False)
```
