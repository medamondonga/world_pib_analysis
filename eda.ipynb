```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pib_data_cleaned.csv")
df.head()
```

## Create and display a small boxplot for each numeric column to visualize its distribution and detect outliers.

```python
my_columns = df.columns.tolist()
for col in my_columns[1:]:
    plt.figure(figsize=(3, 3))
    df.boxplot(column=[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
df.columns = [col.split("_")[1] if "_" in col else col for col in df.columns]
df.head()
```

## Plot the GDP evolution over time for each selected country on a single line chart.

```python

plt.figure(figsize=(12, 6))
annees = df.columns[1:]
countries = df["pays/zone"].unique()
for country in countries:
    subset = df[df["pays/zone"] == country]
    pib_values = subset[annees].values.flatten()   # récupère le PIB sur 1980–2023
    
    plt.plot(annees, pib_values, label=country)

plt.xlabel("Année")
plt.ylabel("PIB")
plt.title("Évolution du PIB par pays (1980–2023)")
plt.legend()
plt.show()
```
